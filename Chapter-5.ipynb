##**Градиентный спуск с несколькими входами**

```
def w_sum(a, b):
  assert(len(a) == len(b))
  output = 0

  for i in range(len(a)):
    output += (a[i] * b[i])
  return output

weights = [0.1, 0.2, -.1]

def neural_network(input, weights):
  pred = w_sum(input, weights)
  return pred

def ele_mul(number, vector):
  output = [0, 0, 0]
  assert(len(output) == len(vector))

  for i in range(len(vector)):
    output[i] = number * vector[i]
  return output

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose = [1, 1, 0, 1]
true = win_or_lose[0]

input = [toes[0], wlrec[0], nfans[0]]
predict = neural_network(input, weights)
error = (predict - true) ** 2
delta = predict - true
weight_deltas = ele_mul(delta, input)

alpha = 0.01

for i in range(len(weights)):
  weights[i] -= alpha * weight_deltas[i]

print("Weights: " + str(weights))
print("Weight_deltas: " + str(weight_deltas))

```

## **Рассмотрим несколько шагов обучения**

```
def neural_network(input, weights):
  out = 0
  for i in range(len(input)):
    out += (input[i] * weights[i])
  return out

def ele_mul(scalar, vector):
  out = [0, 0, 0]
  for i in range(len(out)):
    out[i] = vector[i] * scalar
  return out

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose = [1, 1, 0, 1]
true = win_or_lose[0]

alpha = 0.01
weights = [0.1, 0.2, -.1]
input = [toes[0], wlrec[0], nfans[0]]

for iter in range(3):
  pred = neural_network(input, weights)
  error = (pred - true) ** 2
  delta = pred - true

  weight_deltas = ele_mul(delta, input)

  print("Iteration: " + str(iter+1))
  print("Predict: " + str(pred))
  print("Error: " + str(error))
  print("Delta: " + str(delta))
  print("Weights: " + str(weights))
  print("Weight_deltas: " + str(weight_deltas))
  print("\n##############")

  for i in range(len(weights)):
    weights[i] -= alpha * weight_deltas[i]
```

## **Обучение методом градиентного спуска с несколькими выходами**

```
weights = [0.3, 0.2, 0.9]

def ele_mul(scalar, vector):
  out = [0, 0, 0]
  for i in range(len(out)):
    out[i] = vector[i] * scalar
  return out

def scalar_ele_mul(number, vector):
  out = [0, 0, 0]
  assert(len(out) == len(vector))

  for i in range(len(vector)):
    out[i] = number * vector[i]
  return out

def neural_network(input, weights):
  pred = ele_mul(input, weights)
  return pred

wlrec = [0.65, 0.8, 0.8, 0.9]

hurt = [0.1, 0.0, 0.0, 0.1]
win  = [  1,   1,   0,   1]
sad  = [0.1, 0.0, 0.1, 0.2]

input = wlrec[0]
true = [hurt[0], win[0], sad[0]]

pred = neural_network(input, weights)

error = [0, 0, 0]
delta = [0, 0, 0]

for i in range(len(true)):
  error[i] = (pred[i] - true[i]) ** 2
  delta[i] = pred[i] - true[i]

weight_deltas = scalar_ele_mul(input, weights)
alpha = 0.1

for i in range(len(weights)):
  weights[i] -= (weight_deltas[i] * alpha)

  print("Weights: " + str(weights))
  print("Weights Deltas: " + str(weight_deltas))
  print("###############\n")
```
